```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
if True:
    coverage_path = "results/coverage.csv"
    df = pd.read_csv(coverage_path)
    df["coverage"] = round((df["covered_jumpis"] + 1) / (df["total_jumpis"] + 1) * 100)
    df = df.sort_values(by="bytelen")
    step = round(df.shape[0] / 10)
    xs = []
    ys = []
    for i in range(10):
        start_at = i * step
        end_at = (i + 1) * step
        segment = df[start_at:end_at]
        xs.append(segment["bytelen"].mean())
        ys.append(segment["coverage"].mean())
    plt.figure()
    plt.title("JUMPI Coverage")
    plt.xlabel("Average Bytecode Length (bytes)")
    plt.ylabel("Coverage(%)")
    plt.plot(xs, ys, marker="x")
    plt.show()
```

```python
if True:
    coverage_path = "results/coverage.csv"
    df = pd.read_csv(coverage_path)
    df["coverage"] = round((df["covered_jumpis"] + 1) / (df["total_jumpis"] + 1) * 100)
    xs = [(x + 1) * 10 for x in range(10)]
    ys = []
    total = df.shape[0]
    for x in xs:
        n = df[df["coverage"] >= x].shape[0]
        ys.append(round(n / total * 100))
    plt.figure()
    plt.title("View")
    plt.xlabel("Coverage (%)")
    plt.ylabel("Percentage (%)")
    plt.plot(xs, ys, marker="x")
    plt.show()
```

```python
if True:
    tainting_path = "results/tainting.csv"
    df = pd.read_csv(tainting_path)
    labels = [
        "sloads",
        "mloads",
        "mstores",
        "sstores",
        "failed"
    ]
    plt.figure()
    success_sloads = df["success_sloads"].sum()
    success_mloads = df["success_mloads"].sum()
    success_mstores = df["success_mstores"].sum()
    success_sstores = df["success_sstores"].sum()
    
    failed_sloads = df["failed_sloads"].sum()
    failed_mloads = df["failed_mloads"].sum()
    failed_mstores = df["failed_mstores"].sum()
    failed_sstores = df["failed_sstores"].sum()
    xs = [
        success_sloads,
        success_mloads,
        success_mstores,
        success_sstores,
        failed_sloads + failed_mloads + failed_mstores + failed_sstores
    ]
    plt.pie(xs, labels=labels)
    plt.show()
```
