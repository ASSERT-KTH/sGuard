```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
if True:
    coverage_path = "results/coverage.csv"
    df = pd.read_csv(coverage_path)
    df["coverage"] = round((df["covered_jumpis"] + 1) / (df["total_jumpis"] + 1) * 100)
    df = df.sort_values(by="bytelen")
    step = round(df.shape[0] / 10)
    xs = []
    ys = []
    for i in range(10):
        start_at = i * step
        end_at = (i + 1) * step
        segment = df[start_at:end_at]
        xs.append(segment["bytelen"].mean())
        ys.append(segment["coverage"].mean())
    plt.figure()
    plt.title("JUMPI Coverage")
    plt.xlabel("Average Bytecode Length (bytes)")
    plt.ylabel("Coverage(%)")
    plt.plot(xs, ys, marker="x")
    plt.ylim(0, 100)
    plt.show()
    _coverage = round(df[df["coverage"] == 100].shape[0] / df.shape[0] * 100)
    print("full: " + str(_coverage) + "%")
    _90coverage = round(df[df["coverage"] >= 90].shape[0] / df.shape[0] * 100)
    print("90: " + str(_90coverage) + "%")
    _80coverage = round(df[df["coverage"] >= 80].shape[0] / df.shape[0] * 100)
    print("80: " + str(_80coverage) + "%")
```
